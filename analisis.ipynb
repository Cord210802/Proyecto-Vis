```r
# Cargar librerías necesarias
library(ggplot2)

# Cargar los datos
data <- read.csv("atlas-de-riesgo-inundaciones.csv")

# 1. Gráfico de barras para mostrar la distribución de la intensidad del fenómeno
ggplot(data, aes(x = intensidad)) +
  geom_bar(fill = "skyblue") +
  labs(title = "Distribución de la Intensidad del Fenómeno de Inundación",
       x = "Intensidad",
       y = "Número de Registros") +
  theme_minimal()

# 2. Gráfico de dispersión para mostrar la relación entre alcaldía y área afectada
ggplot(data, aes(x = reorder(alcaldia, area_m2, FUN = median), y = area_m2)) +
  geom_boxplot(fill = "lightblue", outlier.shape = NA) +
  labs(title = "Distribución del Área Afectada por Inundaciones por Alcaldía",
       x = "Alcaldía",
       y = "Área Afectada (m²)") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_y_log10(labels = scales::comma) +
  coord_flip()
```
