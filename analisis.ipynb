```r
# Cargar libreras necesarias
library(ggplot2)

# Cargar los datos
data <- read.csv("atlas-de-riesgo-inundaciones.csv")

# 1. Grfico de barras para mostrar la distribucin de la intensidad del fenmeno
ggplot(data, aes(x = intensidad)) +
  geom_bar(fill = "skyblue") +
  labs(title = "Distribucin de la Intensidad del Fenmeno de Inundacin",
       x = "Intensidad",
       y = "Nmero de Registros") +
  theme_minimal()

# 2. Grfico de dispersin para mostrar la relacin entre alcalda y rea afectada
ggplot(data, aes(x = reorder(alcaldia, area_m2, FUN = median), y = area_m2)) +
  geom_boxplot(fill = "lightblue", outlier.shape = NA) +
  labs(title = "Distribucin del rea Afectada por Inundaciones por Alcalda",
       x = "Alcalda",
       y = "rea Afectada (m)") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_y_log10(labels = scales::comma) +
  coord_flip()
```
